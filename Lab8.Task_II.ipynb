# Task II

## Problem statement

Застосувати синтаксичні залежності для визначення намірів.

## Presets

```python
import spacy
import json

with open("taxiandcars.json") as f:
    data = json.load(f)

utterances = [turn['utterance'] for dialogue in data for turn in dialogue['turns'] if turn['speaker'] == 'USER']

nlp = spacy.load("en_core_web_sm")

sentences = '.'.join(utterances)

doc = nlp(sentences)
```

## Solution

```python
for token in doc:
    if token.dep_ == "dobj" and token.text.lower() in ('car', 'taxi', 'cab', 'ride','machine', 'automobile', 'vehicle', 'carriage'):
        print('Intent: ', token.head.text + token.text.capitalize())
```
