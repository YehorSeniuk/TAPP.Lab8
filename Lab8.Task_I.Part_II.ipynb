# Task I. Presets

```python
import spacy
from spacy.matcher import Matcher
import json

with open("taxiandcars.json") as f:
    data = json.load(f)

utterances = [turn['utterance'] for dialogue in data for turn in dialogue['turns']]

nlp = spacy.load("en_core_web_sm")

text = '.'.join(utterances)
```

# Task I. Part II

## Problem statement

Виділити висловлювання користувача, що є підтвердженнями (наприклад, Yeah, that sounds good),  за допомогою шаблонів.

## Solution

```python
matcher = Matcher(nlp.vocab)

pattern = [{"LOWER": {"IN": ["yeah", "yes", "sure", "okay", "ok", "alright", "yep", "exactly", "absolutely", "good", "fine"]}}]
matcher.add("Confirmation", [pattern])

doc = nlp(text)

matches = matcher(doc)

for match_id, start, end in matches:
    span = doc[start:end]
    print('Confirmation:', span.text)
```
